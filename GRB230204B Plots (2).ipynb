```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
times = np.array([5687,6004,6332,6659,6981,7310,7633,7954,8275,8598])
m_grb = np.array([15.4379, 15.5705, 15.7620, 15.8256, 15.9086, 16.0401, 16.1817, 16.2772, 16.3809, 16.4301])

m,c = np.polyfit(times, m_grb, 1)

y_fit = m*times+c

print(m)

plt.title('Apparent magnitude of GRB230204B as a function of time')
plt.xlabel('Time since T0 / s')
plt.ylabel('Apparent magnitude')
plt.scatter(times, m_grb)
plt.plot(times, y_fit, color='red')
plt.ylim(16.5,15.4)
plt.xlim(5000,9000)
plt.show()
```

```python
times = np.array([5687,6004,6332,6659,6981,7310,7633,7954,8275,8598])
m_grb = np.array([15.4379, 15.5705, 15.7620, 15.8256, 15.9086, 16.0401, 16.1817, 16.2772, 16.3809, 16.4301])

err_zp = np.array([0.0355, 0.0355, 0.0492, 0.0488, 0.0316, 0.0393, 0.0248, 0.0428, 0.0217, 0.0272])

err_inst = np.array([0.0161, 0.0188, 0.0171, 0.0275, 0.0345, 0.0295, 0.0256, 0.0537, 0.0364, 0.0440])
err_fin = np.sqrt((err_zp)**2 + err_inst**2)

logtime = np.log10(times)

m,c = np.polyfit(logtime, m_grb, 1)

y_fit = m * logtime + c

print(m/2.5)


plt.title('Apparent magnitude of GRB230204B as a function of time')
plt.xlabel('Time since T0 / s')
plt.ylabel('Apparent magnitude')
plt.scatter(logtime, m_grb)
plt.plot(logtime, y_fit, color='red')
plt.ylim(16.5,15.4)


plt.errorbar(logtime, m_grb, yerr=err_fin, fmt='o', capsize=5, label='Symmetric Errors')

plt.savefig(f'/Users/aaravrastogi/Downloads/Apparent magnitude of GRB230204B')

plt.show()

```
